# Importing Modules

Firstly lets import some important modules to get started 

```python
import pygame
```

#### pygame.init() 

`pygame.init()` => intializes all the modules for the pyame to work perfectly fine.

#### pygame.time

`pygame.time` => used for monitoring time
- `pygame.time.Clock()` => create an object to track time
- - `pygame.time.Clock.ticl` => upadte the clock 

#### pyame.font

`pygame.font` => used for fonts
- `pygame.font.Font` => used to create a new font from exteranl file
1. `filename` => name of the file 
2. `pathlib` => path to the file 
3. `object` => object as the font
4. `size` => size of the font 
- - `pygame.font.Font.render` => draw text on a new surface
- 1. `text` => text to be renderd
- 2. `antialias` => wether the characyers to be having smooth edges or not 
- 3. `color` => color of the text
- 4. `background` => color of the background of the text


#### pygame.display

`pygame.display` => used to display the central window and screen 
- `pygame.display.set_mode()` => Intitalize the window
1. `size` = pair of numbers representing height and width 
2. `flags` = shows which type of display you want 
3. `depth` = the amound of depth into the window
4. `display` = type of display 
5. `vsync` = turning on the vertical sync or not 
- - `pygame.display.setmode().blit` => draws a source surface onto the main surface
- 1. `source` => draws a surface onto the main surface
- 2. `dest` = positioning of the surface
- 3. `area` => area of the surface to be used
- - `pygame.display.setmode().fill` => fills the surface with the colour 

#### pygame.image

`pygame.image` => module for image processing
- `pygame.image.load` => load a new file as image 
1. `filename` => name of the file to import 
2. `fileobj` => import image as object 

#### pygame.transform

`pygame.transform` => used to tranform the surafces
- `pygame.transform.scale` => used to resize the scale of the object 
1. `surface` => name of the surface
2. `size` => pair of height and width 
- `pygame.transform.flip` => flip vertically or horizontally
1. `surface` => name of the surface
2. `flip_x` => if True, flips vertically
3. `flip_y` => if True, flips horizontally
- `pygame.tranform.rotate` => rotate the image
1. `surface` => name of the surface 
2. `angle` => angle at which to be rotated

#### pygame.mixer

`pygame.mixer` => module for sound processing
- `pygame.mixer.init`=> used to intialize the mixer module 
1. `frquency` => frequncy of the audio file
2. `size` => shows how many bits are used for each audio sample
3. `channels` => wether to use mono or sterie audio system
4. `buffer` => the number of internal samples used in the sound mixer
5. `devicename` => name of the output device for the pygame mixer 
6. `allowedchanges` => pending
- `pygame.mixer.music` => used for controlling streamed audio
- - `pygame.mixer.music.load` => loading a music file for playback
- 1. `filename` => name of the file 
- 2. `fileobj` => audio file as an object
- - `pygame.mixer.music.play` => start the playback of the sound
- 1. `loops` => number of loops for the sound to be played
- 2. `start` => start from which position
- 3. `fade_ms` => timeline at which the sound will start fading in
- `pygame.mixer.Sound` => used for creating new sound object from file 
1. `filename` => name of the file 
2. `buffer` => number of internal sound samples
3. `object` => sound as object
4. `array` => sound as an array 
- - `pygame.mixer.Sound.set_volume` => sets the volume for the sound


#### pygame.mouse

`pygame.mouse` => to work with the mouse module
- `pygame.mouse.get_pos` => get the mouse cursor position
- `pygame.mouse.get_pressed` => get the state of the mouse buttons



#### pygame.draw

`pygame.draw` => used for drawing shapes
- `pygame.draw.circle` => draw a circle
1. `surface` => surafce to be used for drawing circle 
2. `color` => color of the circle 
3. `center` => center point of the circle 
4. `radius` => radius of the circle
5. `width` => thickness of the cricle 
6. `draw_top_right` => if True, then top right of the circle will be filled 
7. `draw_top_left` => if True, then top left of the circle will be filled
8. `draw_bottom_left` => if True, then bottom left of the circle will be filled 
9. `draw_bottom_right` => if True, then bottom right of the circle will be filled 
- `pygame.draw.rect` => draw a rectangle 
1. `surface` => name of the surface to be drawned on the rectangle
2. `color` => color of the rectangle 
3. `rect` => dimensions/coordinates of the rectangle
4. `width` => widht of the rectangle \
5. `border_radius` => radius of the border
6. `border_top_left_radius` => radius of the border of the top left corner 
7. `border_top_right_radius` => radius of the border of the top right corner 
8. `border_bottom_left_radius` => radius of the border of the bottom left corner 
9. `border_bottom_right_radius` => radius of the border of the bottom right corner

```python
# Python Arcade Shooting Gallery!
import pygame
import math

pygame.init()

fps = 60
width = 900
height = 800
level = 0
points = 0
total_shots = 0
mode = 0
ammo = 0
time_passed = 0
time_remaining = 0
best_freeplay = 0
best_ammo = 0
best_timed = 0
counter = 1

bgs = []
banners = []
guns = []

target_images = [[], [], []]
one_coords = [[], [], []]
two_coords = [[], [], []]
three_coords = [[], [], [], []]

targets = {1: [10, 5, 3],
           2: [12, 8, 5],
           3: [15, 12, 8, 3]}

shot = False
game_over = False
pause = False
write_values = False
clicked = False

menu = True
new_coords = True

timer = pygame.time.Clock()

font = pygame.font.Font('assets/font/myFont.ttf', 32)
big_font = pygame.font.Font('assets/font/myFont.ttf', 60)

screen = pygame.display.set_mode([width, height])

menu_img = pygame.image.load(f'assets/menus/mainMenu.png')
game_over_img = pygame.image.load(f'assets/menus/gameOver.png')
pause_img = pygame.image.load(f'assets/menus/pause.png')

for i in range(1, 4):

    bgs.append(pygame.image.load(f'assets/bgs/{i}.png'))
    banners.append(pygame.image.load(f'assets/banners/{i}.png'))
    guns.append(pygame.transform.scale(pygame.image.load(f'assets/guns/{i}.png'), (100, 100)))

    if i < 3:

        for j in range(1, 4):

            target_images[i - 1].append(pygame.transform.scale(pygame.image.load(f'assets/targets/{i}/{j}.png'), (120 - (j * 18), 80 - (j * 12))))
    
    else:

        for j in range(1, 5):
    
            target_images[i - 1].append(pygame.transform.scale(pygame.image.load(f'assets/targets/{i}/{j}.png'), (120 - (j * 18), 80 - (j * 12))))

file = open('high_scores.txt', 'r')
read_file = file.readlines()
file.close()

best_freeplay = int(read_file[0])
best_ammo = int(read_file[1])
best_timed = int(read_file[2])

pygame.mixer.init()
pygame.mixer.music.load('assets/sounds/bg_music.mp3')

plate_sound = pygame.mixer.Sound('assets/sounds/Broken plates.wav')
bird_sound = pygame.mixer.Sound('assets/sounds/Drill Gear.mp3')
laser_sound = pygame.mixer.Sound('assets/sounds/Laser Gun.wav')

plate_sound.set_volume(.2)
bird_sound.set_volume(.2)
laser_sound.set_volume(.3)

pygame.mixer.music.play()

def draw_score():

    points_text = font.render(f'Points: {points}', True, 'black')
    shots_text = font.render(f'Total Shots: {total_shots}', True, 'black')
    time_text = font.render(f'Time Elapsed: {time_passed}', True, 'black')

    screen.blit(points_text, (320, 660))
    screen.blit(shots_text, (320, 687))
    screen.blit(time_text, (320, 714))

    if mode == 0:

        mode_text = font.render(f'Freeplay!', True, 'black')

    if mode == 1:

        mode_text = font.render(f'Ammo Remaining: {ammo}', True, 'black')

    if mode == 2:

        mode_text = font.render(f'Time Remaining {time_remaining}', True, 'black')

    screen.blit(mode_text, (320, 741))


def draw_gun():

    mouse_pos = pygame.mouse.get_pos()
    gun_point = (width / 2, height - 200)
    lasers = ['red', 'purple', 'green']
    clicks = pygame.mouse.get_pressed()
    
    if mouse_pos[0] != gun_point[0]:
    
        slope = (mouse_pos[1] - gun_point[1]) / (mouse_pos[0] - gun_point[0])
    
    else:
    
        slope = -100000
    
    angle = math.atan(slope)
    rotation = math.degrees(angle)
    
    if mouse_pos[0] < width / 2:
    
        gun = pygame.transform.flip(guns[level - 1], True, False)
    
        if mouse_pos[1] < 600:
    
            screen.blit(pygame.transform.rotate(gun, 90 - rotation), (width / 2 - 90, height - 250))
    
            if clicks[0]:
    
                pygame.draw.circle(screen, lasers[level - 1], mouse_pos, 5)
    
    else:
    
        gun = guns[level - 1]
    
        if mouse_pos[1] < 600:
    
            screen.blit(pygame.transform.rotate(gun, 270 - rotation), (width / 2 - 30, height - 250))
    
            if clicks[0]:
    
                pygame.draw.circle(screen, lasers[level - 1], mouse_pos, 5)

def move_level(coords):

    if level == 1 or level == 2:

        max_val = 3

    else:

        max_val = 4

    for i in range(max_val):

        for j in range(len(coords[i])):

            my_coords = coords[i][j]

            if my_coords[0] < -150:

                coords[i][j] = (width, my_coords[1])

            else:

                coords[i][j] = (my_coords[0] - 2 ** i, my_coords[1])

    return coords

def draw_level(coords):

    if level == 1 or level == 2:

        target_rects = [[], [], []]

    else:

        target_rects = [[], [], [], []]

    for i in range(len(coords)):

        for j in range(len(coords[i])):

            target_rects[i].append(pygame.rect.Rect((coords[i][j][0] + 20, coords[i][j][1]),(60 - i * 12, 60 - i * 12)))
            screen.blit(target_images[level - 1][i], coords[i][j])
    
    return target_rects

def check_shot(targets, coords):

    global points

    mouse_pos = pygame.mouse.get_pos()

    for i in range(len(targets)):

        for j in range(len(targets[i])):

            if targets[i][j].collidepoint(mouse_pos):

                coords[i].pop(j)
                points += 10 + 10 * (i ** 2)

                if level == 1:

                    bird_sound.play()

                elif level == 2:

                    plate_sound.play()

                elif level == 3:

                    laser_sound.play()

    return coords


def draw_menu():

    global game_over, pause, mode, level, menu, time_passed, total_shots, points, ammo, time_remaining, best_freeplay, best_ammo, best_timed, write_values, clicked, new_coords 
    
    game_over = False
    pause = False
    
    screen.blit(menu_img, (0, 0))
    mouse_pos = pygame.mouse.get_pos()
    clicks = pygame.mouse.get_pressed()
    
    freeplay_button = pygame.rect.Rect((170, 524), (260, 100))
    ammo_button = pygame.rect.Rect((475, 524), (260, 100))
    timed_button = pygame.rect.Rect((170, 661), (260, 100))
    reset_button = pygame.rect.Rect((475, 661), (260, 100))
    
    screen.blit(font.render(f'{best_freeplay}', True, 'black'), (340, 580))
    screen.blit(font.render(f'{best_ammo}', True, 'black'), (650, 580))
    screen.blit(font.render(f'{best_timed}', True, 'black'), (350, 710))
    
    if freeplay_button.collidepoint(mouse_pos) and clicks[0] and not clicked:
    
        mode = 0
        time_passed = 0
        points = 0
        total_shots = 0
        level = 1
        menu = False
        clicked = True
        new_coords = True
    
    if ammo_button.collidepoint(mouse_pos) and clicks[0] and not clicked:
        time_passed = 0
        total_shots = 0
        points = 0
        mode = 1
        level = 1
        ammo = 81
        menu = False
        clicked = True
        new_coords = True
    
    if timed_button.collidepoint(mouse_pos) and clicks[0] and not clicked:
    
        time_passed = 0
        total_shots = 0
        points = 0
        level = 1
        mode = 2
        time_remaining = 30
        menu = False
        clicked = True
        new_coords = True
    
    if reset_button.collidepoint(mouse_pos) and clicks[0] and not clicked:
    
        best_freeplay = 0
        best_ammo = 0
        best_timed = 0
        clicked = True
        write_values = True

def draw_game_over():

    global clicked, level, pause, game_over, menu, points, total_shots, time_passed, time_remaining

    if mode == 0:

        display_score = time_passed

    else:

        display_score = points

    screen.blit(game_over_img, (0, 0))

    mouse_pos = pygame.mouse.get_pos()
    clicks = pygame.mouse.get_pressed()

    exit_button = pygame.rect.Rect((170, 661), (260, 100))
    menu_button = pygame.rect.Rect((475, 661), (260, 100))

    screen.blit(big_font.render(f'{display_score}', True, 'black'), (650, 570))

    if menu_button.collidepoint(mouse_pos) and clicks[0] and not clicked:

        level = 0
        points = 0
        total_shots = 0
        time_passed = 0
        time_remaining = 0
        pause = False
        game_over = False
        menu = True
        clicked = True

    if exit_button.collidepoint(mouse_pos) and clicks[0] and not clicked:

        global run

        run = False

def draw_pause():

    global level, pause, menu, points, total_shots, time_passed, time_remaining, clicked, new_coords

    screen.blit(pause_img, (0, 0))

    mouse_pos = pygame.mouse.get_pos()
    clicks = pygame.mouse.get_pressed()

    resume_button = pygame.rect.Rect((170, 661), (260, 100))
    menu_button = pygame.rect.Rect((475, 661), (260, 100))

    if resume_button.collidepoint(mouse_pos) and clicks[0] and not clicked:

        level = resume_level
        pause = False
        clicked = True

    if menu_button.collidepoint(mouse_pos) and clicks[0] and not clicked:

        pygame.mixer.music.play()

        level = 0
        points = 0
        total_shots = 0
        time_passed = 0
        time_remaining = 0
        pause = False
        menu = True
        clicked = True
        new_coords = True

run = True

while run:

    timer.tick(fps)

    if level != 0:

        if counter < 60:

            counter += 1

        else:

            counter = 1
            time_passed += 1

            if mode == 2:

                time_remaining -= 1

    if new_coords:

        one_coords = [[], [], []]
        two_coords = [[], [], []]
        three_coords = [[], [], [], []]
        
        for i in range(3):
        
            my_list = targets[1]
        
            for j in range(my_list[i]):
        
                one_coords[i].append((width // (my_list[i]) * j, 300 - (i * 150) + 30 * (j % 2)))
        
        for i in range(3):
        
            my_list = targets[2]
        
            for j in range(my_list[i]):
        
                two_coords[i].append((width // (my_list[i]) * j, 300 - (i * 150) + 30 * (j % 2)))
        
        for i in range(4):
        
            my_list = targets[3]
        
            for j in range(my_list[i]):
        
                three_coords[i].append((width // (my_list[i]) * j, 300 - (i * 100) + 30 * (j % 2)))
        
        new_coords = False

    screen.fill('black')
    screen.blit(bgs[level - 1], (0, 0))
    screen.blit(banners[level - 1], (0, height - 200))
    
    if menu:
    
        level = 0
    
        draw_menu()
    
    if game_over:
    
        level = 0
    
        draw_game_over()
    
    if pause:
    
        level = 0
    
        draw_pause()

    
    if level == 1:
    
        target_boxes = draw_level(one_coords)
        one_coords = move_level(one_coords)
    
        if shot:
    
            one_coords = check_shot(target_boxes, one_coords)
            shot = False
    
    elif level == 2:
    
        target_boxes = draw_level(two_coords)
        two_coords = move_level(two_coords)
    
        if shot:
    
            two_coords = check_shot(target_boxes, two_coords)
            shot = False
    
    elif level == 3:
    
        target_boxes = draw_level(three_coords)
        three_coords = move_level(three_coords)
    
        if shot:
    
            three_coords = check_shot(target_boxes, three_coords)
            shot = False
    
    if level > 0:
    
        draw_gun()
        draw_score()

    
    for event in pygame.event.get():
    
        if event.type == pygame.QUIT:
    
            run = False
    
        if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:
    
            mouse_position = pygame.mouse.get_pos()
    
            if (0 < mouse_position[0] < width) and (0 < mouse_position[1] < height - 200):
    
                shot = True
                total_shots += 1
    
                if mode == 1:
    
                    ammo -= 1
    
            if (670 < mouse_position[0] < 860) and (660 < mouse_position[1] < 715):
    
                resume_level = level
                pause = True
                clicked = True
    
            if (670 < mouse_position[0] < 860) and (715 < mouse_position[1] < 760):
    
                menu = True
                pygame.mixer.music.play()
                clicked = True
                new_coords = True
    
        if event.type == pygame.MOUSEBUTTONUP and event.button == 1 and clicked:
    
            clicked = False

    if level > 0:
    
        if target_boxes == [[], [], []] and level < 3:
    
            level += 1
    
        if (level == 3 and target_boxes == [[], [], [], []]) or (mode == 1 and ammo == 0) or (
    
                mode == 2 and time_remaining == 0):
    
            new_coords = True
    
            pygame.mixer.music.play()
    
            if mode == 0:
    
                if time_passed < best_freeplay or best_freeplay == 0:
    
                    best_freeplay = time_passed
                    write_values = True
    
            if mode == 1:
    
                if points > best_ammo:
    
                    best_ammo = points
                    write_values = True
    
            if mode == 2:
    
                if points > best_timed:
    
                    best_timed = points
                    write_values = True
    
            game_over = True
    
    if write_values:
    
        file = open('high_scores.txt', 'w')
        file.write(f'{best_freeplay}\n{best_ammo}\n{best_timed}')
        file.close()
    
        write_values = False
    
    pygame.display.flip()

pygame.quit()
```
